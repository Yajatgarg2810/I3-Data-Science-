```python
# Q1.

# Input string
s = 'Python is awesome and fun'

# Variables to keep track of the longest word length
max_length = 0
current_length = 0

# Loop through each character in the string
for char in s:
    if char != ' ':  # If the character is not a space, increase current length
        current_length += 1
    else:  # If it is a space, compare and reset
        if current_length > max_length:
            max_length = current_length
        current_length = 0

# Final check for the last word
if current_length > max_length:
    max_length = current_length

# Output the length of the longest word
print(max_length)
```

```python
# Q2.

# Input string
s = 'apple pie and apple juice'
target = 'apple'
replacement = 'oranges'

# Variables
result = ""
i = 0

# Loop through the string
while i < len(s):
    # Check if the target word matches starting at the current index
    if s[i:i+len(target)] == target:
        # Add the replacement to the result
        result += replacement
        # Skip the length of the target word
        i += len(target)
    else:
        # Add the current character to the result
        result += s[i]
        i += 1

# Output the result
print(result)
```

```python
# Q3.

# Strings to compare
s1 = 'Hello'
s2 = 'hello'

# Check if strings are equal ignoring case
if len(s1) != len(s2):  # First, ensure the lengths are equal
    print("Strings are not equal")
else:
    is_equal = True
    for i in range(len(s1)):
        # Compare characters manually after converting to the same case
        c1 = s1[i]
        c2 = s2[i]

        # Convert to lowercase (manual normalization)
        if 'A' <= c1 <= 'Z':  # If c1 is uppercase
            c1 = chr(ord(c1) + 32)
        if 'A' <= c2 <= 'Z':  # If c2 is uppercase
            c2 = chr(ord(c2) + 32)

        # Compare the normalized characters
        if c1 != c2:
            is_equal = False
            break

    if is_equal:
        print("Strings are equal")
    else:
        print("Strings are not equal")
```

```python
# Q4.

# Input string
s = 'abcdefgh'

# Initialize result
result = ""

# Loop through the string
for i in range(len(s)):
    # Check if the index is odd (every second character)
    if i % 2 == 1:
        result += s[i]

# Output the result
print(result)
```

```python
# Q5.

# Input string
s = "12345"

# Flag to indicate if the string contains only numeric digits
is_numeric = True

# Loop through each character in the string
for char in s:
    if not ('0' <= char <= '9'):  # Check if the character is outside the numeric range
        is_numeric = False
        break

# Output the result
if is_numeric:
    print("The string contains only numeric digits.")
else:
    print("The string contains non-numeric characters.")
```

```python
# Q6.

# Input string
s = 'python is fun'

# Variables
result = ""
capitalize_next = True  # Flag to determine if the next character should be capitalized

# Loop through each character in the string
for char in s:
    if char == ' ':  # If the character is a space
        result += char
        capitalize_next = True  # Set the flag to capitalize the next letter
    elif capitalize_next:  # If we need to capitalize the current character
        if 'a' <= char <= 'z':  # Check if it's a lowercase letter
            result += chr(ord(char) - 32)  # Convert to uppercase
        else:
            result += char
        capitalize_next = False  # Reset the flag
    else:  # For non-capitalized characters
        result += char

# Output the result
print(result)
```
